```python
import gradio as gr
import cv2 as cv
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('main_model.h5')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('ArSL_Data_Labels.csv')
file
```

```python
class_names = file['Class'].to_list()
class_set = pd.unique(class_names).tolist()
class_set
```

```python
def preprocess(img):
   img = img.convert("L").resize((64, 64))
   img = np.array(img) / 255.0
   img = np.expand_dims(img, axis=-1)
   img = np.expand_dims(img, axis=0)
   return img

def predict_sign(img):
    processed_img = preprocess(img)
    preds = model.predict(processed_img)
    predicted_index = np.argmax(preds)
    return class_set[predicted_index]
```

```python
import PIL.Image as Image
```

```python
img = Image.open("KHAA (2).JPG").convert("L").resize((64, 64))
img = np.array(m) / 255.0

img_array = np.expand_dims(m, axis=-1)   # (64,64,1)
img_array = np.expand_dims(img_array, axis=0)  # (1,64,64,1)

pred = model.predict(img_array)
print("Predicted index:", np.argmax(pred))
```

```python
interface = gr.Interface(
    fn=predict_sign,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="Sign Language Recognizer",
    description="Upload a hand sign image and the model will predict the corresponding character."
)
interface.launch(share = True)
```
